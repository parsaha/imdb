# IMDb Exploratory Data Analysis

The overall goal of the project is to create a notebook that will allow the user to see average IMDb ratings for a given actor. To achieve this goal, we need to first understand what the data is, where to get it from, how to get it, and so on. 

Since I have no knowledge of web scraping prior to this, my main reference on how to proceed finding and understanding the data is [this site](https://www.dataquest.io/blog/web-scraping-beautifulsoup/). 

First, we import our packages. 

```python
from matplotlib import pyplot as plt
%matplotlib inline
import pandas as pd
from bs4 import BeautifulSoup
from statistics import mean
```

Now, we use [Zach Galifinakais](https://www.imdb.com/name/nm0302108/) as our lab rat (henceforth referred to as simply "Zach"). We will use his IMDb page to understand the data. 

__Note__: the link provided (which you would likely find if you were to simply look up his IMDb page) is different from the link we are exploring; this can be easily addressed by adding '#actor' to the end of the generic link above. 

Below, we will see the first 500 characters of of the html for Zach's page:

```python
from requests import get
url = 'https://www.imdb.com/name/nm0302108/#actor'
response = get(url)
print(response.text[:500])
```

```python
html_soup = BeautifulSoup(response.text, 'html.parser')
type(html_soup)
```

```python
movie_containers = html_soup.find_all('div', 'filmo-category-section')
print(type(movie_containers))
print(len(movie_containers))
```

```python
print(movie_containers[0]) 
```

The 9 are different categories of contrbution (producer, etc).
So we take the first one (which happens to be Actor)

Since we realize that the only thing we care about is the first element, let's just make our variable that first element:

```python
movie_containers = movie_containers[0]
```

```python
print(type(movie_containers))
print(len(movie_containers))
```

Taking a look at the data, I saw that the years show up in the `span` classes of the html of `movie_containers`:

```python
all_years = movie_containers.find_all('span')
```

To show a sample of the elements of `all_years`, I print the first element of `all_years`. 

```python
str(all_years[0])
```

Now we isolate the actual year portion of `all_years[0]`:

```python
str(all_years[0]).split('\n')[1][1:]
```

Since that is the only part we care about, we can replace `all_years` with a list of the isolated years. 

```python
year_list = []
for year in all_years:
    year_list.append(str(year).split('\n')[1][1:])
```

As we saw with our example above, some year entries are ranges; this is because some of Zach's roles were in television, whose years are the timeframe for the show. Since we are dealing with ratings and ratings for television shows are most likely given based on the most recent iteration of the show, we will make the year entry of the shows their most recent year. For example, we will make `all_years[0]` 2019 instead of 2016-2019. 

Additionally, some works that share their names with other works are appended with a Roman Numeral (i.e. I, II, etc.) to differentiate them; therefore, we also remove these numerals in the code below. 

```python
# makes ranges of years (like for shows) return the most recent year (since ratings are usually for most recent version of show)
# and then makes the strings integers. 
print(year_list, '\n')
for index, year in enumerate(year_list):
    if '-' in year:
        year_list[index] = year.split('-')[1]
    if '/I' in year:
        year_list[index] = year.split('/I')[0]
    year_list[index] = int(year_list[index])
print(year_list)
```

Now we move on to finding the movie titles. 

```python
movie_list = movie_containers.find_all('b')
```

The five latest movies are:

```python
movie_list[:5]
```

As an example, we will look at the second element of this list: `movie_list[1]`, which is called The Missing Link, and retrieve its name:

```python
print(movie_list[1].contents[0].contents[0])
```

```python
print(str(movie_list[1].find('a')))
```

Below, we find a way to extract the title and IMDb tag of the movies.

```python
print(str(movie_list[1].find('a')).split('>')[-2][:-3])
print(str(movie_list[1].find('a')).split('/')[2])
```

And now we store them in lists. 

```python
title_list = []
tag_list = []
for item in movie_list:
    title_list.append(str(item.find('a')).split('>')[-2][:-3])
    tag_list.append(str(item.find('a')).split('/')[2])
```

```python
print(len(title_list))
print(len(tag_list))
```

```python
print(title_list[:5])
```

Now, we have a list of all of Zach's movies and their respective IMDb tags (to find their pages). 

Case in point, here are the first 1000 characters of the html for his film Baskets (whose tag, tt6348138, I've just copied and pasted in the appropriate place in `example_url`):

```python
example_url = 'https://www.imdb.com/title/' + 'tt6348138' + '/'
example_response = get(example_url)
print(example_response.text[:1000])
```

After a bit of digging, I found the way to obtain the ratings for each movie's page. Below I add them to a `rating_list` as decimals. 

```python
rating_list = []
for tag in tag_list:
    url = 'https://www.imdb.com/title/' + tag + '/'
    response = get(url)
    html_soup = BeautifulSoup(response.text, 'html.parser')
    rating_containers = html_soup.find_all('div','ratingValue')
    rating_str = str(rating_containers[0].find_all('strong')[0])
    rating = float(rating_str.split(' based on ')[0][-3:])
    rating_list.append(rating)
```

Since the tags are being iterated through in the same order as the movies are listed, the positions of the rating for each movie are preserved. 

```python
print(rating_list)
```

```python
mean(rating_list)
```

Now, some basic visualization that also displays how each year stands against the average.

```python
plt.subplot(1,2,1)
plt.xlabel('Year')
plt.ylabel('Rating')
plt.scatter(year_list,rating_list)
plt.plot([1995, 2020], [mean(rating_list), mean(rating_list)], label='Average', c='r')
plt.legend()
plt.tight_layout()

plt.subplot(1,2,2)
plt.xlabel('Year')
plt.ylabel('Rating')
plt.bar(year_list, rating_list)
plt.plot([1995, 2020], [mean(rating_list), mean(rating_list)], label='Average', c='r')
plt.legend()
plt.tight_layout()
```

To make the data readable, here are the main things we've found in a pandas DataFrame. 

```python
df = pd.DataFrame(list(zip(rating_list, year_list, title_list)), columns=['Ratings', 'Year', 'Title'])
```
